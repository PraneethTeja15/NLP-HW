```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

# ---------------------------------------------------
# Paragraph used for tokenization
# ---------------------------------------------------
text = """NLP is transforming the world. It's used in chatbots, translation, and search engines.
Many students don't realize how powerful language models are.
Today, AI-driven tools are everywhere."""

print("\nORIGINAL PARAGRAPH:\n")
print(text)

# ---------------------------------------------------
# 1. Naive space-based tokenization
# ---------------------------------------------------
naive_tokens = text.split()

print("\nNaive Tokenization (space-based):\n")
print(naive_tokens)

# ---------------------------------------------------
# 2. Manual tokenization (handling punctuation,
#    clitics, suffixes, hyphens)
# ---------------------------------------------------
manual_tokens = [
'NLP','is','transforming','the','world','.',
'It',"'s",'used','in','chatbots',',','translation',',',
'and','search','engines','.',
'Many','students','do',"n't",'realize','how','powerful',
'language','models','are','.',
'Today',',','AI','-','driven','tools','are','everywhere','.'
]

print("\nManual Corrected Tokens:\n")
print(manual_tokens)

# ---------------------------------------------------
# 3. NLTK tokenizer output
# ---------------------------------------------------
nltk_tokens = word_tokenize(text)

print("\nNLTK Tokenization:\n")
print(nltk_tokens)

# ---------------------------------------------------
# 4. Compare differences
# ---------------------------------------------------
print("\nTokens in manual but different from NLTK:\n")

for m, n in zip(manual_tokens, nltk_tokens):
    if m != n:
        print(f"Manual: {m}  |  NLTK: {n}")

print("\nTotal tokens:")
print("Naive:", len(naive_tokens))
print("Manual:", len(manual_tokens))
print("NLTK:", len(nltk_tokens))
```
